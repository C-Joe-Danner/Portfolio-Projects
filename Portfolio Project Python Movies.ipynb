```python
# Importing all the packages that will be needed

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Reading the Data

df = pd.read_csv(r'C:\Users\Joe\Desktop\Joe\Career\Data Analysts\Portfolio\Project 4 Correlation in Python\movies.csv')
```

```python
df
```

```python
# Finding Absent Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Column Data Types

print(df.dtypes)
```

```python
# Dropping Duplicates

df.drop_duplicates()
```

```python
# Data Sorting

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Budget vs Gross Scatter Plot

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget of Movie')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Budget vs Gross Regression Plot 

sns.regplot(x='budget', y='gross', data=df)
```

```python
# Correlation by Method
#Pearson

df.corr(method='pearson')
```

```python
# Kendall

df.corr(method='kendall')
```

```python
# Spearman

df.corr(method='spearman')
```

```python
# Correlation Heatmap

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")

plt.show()
```

```python
# Factorize

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
# Heatmap for Factorized Correlation

correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Movies")

plt.show()
```

```python
# Correlation Pairs Unstacked

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
#Highest Correlation Pairs

sorted_pairs = corr_pairs.sort_values(kind="quicksort")
high_corr = sorted_pairs[abs(sorted_pairs) > 0.5]

print(high_corr)
```
